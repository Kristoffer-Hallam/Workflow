```python
import numpy as np
import matplotlib.pyplot as plt
import pyshtools as pysh
from pyshtools import constant
import numpy.testing as npt
import pandas as pd
```

```python
a = constant.Earth.a_wgs84.value
f = constant.Earth.f_wgs84.value
print(a,f)
```

```python
infile = '../data/EIGEN-6C4_mod.txt'
clm = pysh.SHCoeffs.from_file(infile)
grav = clm.expand()#(a=a, f=f)
```

```python
grav = clm.expand(lmax=95)
fig2, ax2 = grav.plot(cmap='RdBu_r',show=False)
```

```python
import pygmt

fig = pygmt.Figure()
# Use the ISO country code for Brazil and add a padding of 2 degrees (+R2)
fig, ax = grav.plot(cmap='RdBu_r',show=False)
fig.coast(
    projection="B-55/-15/-25/0/8i", region="BR+R2", frame="afg", land="gray", borders=1
)
fig.show()
```
